```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('telecom.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis = 1,inplace = True)
```

```python
df.head()
```

```python
x = df.drop('Churn',axis = 1)
x.head()
```

```python
y = df['Churn']
y.head()
```

### scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
x_train_transform = scaler.fit_transform(x_train)
x_test_transfrom = scaler.fit_transform(x_test)
```

```python
y_train_transform = scaler.fit_transform(pd.DataFrame(y_train))
y_test_transform = scaler.fit_transform(pd.DataFrame(y_test))
```

### decision tree classifier

```python
DecisionTree = DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
DecisionTree.fit(x_train_transform,y_train)
```

```python
DecisionTree.score(x_test_transfrom,y_test)
```

```python
y_pred = DecisionTree.predict(x_test_transfrom)
y_pred
```

### classification report

```python
print(classification_report(y_test,y_pred))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.2)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()#sampling_strategy='minority', random_state=42
x_train_resampled, y_train_resampled = sm.fit_resample(x_train,y_train)
```

```python
y_train_resampled.value_counts()
```

```python
DecisionClassifer_smote = DecisionTreeClassifier(criterion = 'gini',max_depth = 8)
```

```python
DecisionClassifer_smote.fit(x_train_resampled,y_train_resampled)
```

```python
y_pred = DecisionClassifer_smote.predict(x_test)
model_score = DecisionClassifer_smote.score(x_test,y_test)
print(model_score)
print(metrics.classification_report(y_pred,y_test))
```

```python
print(metrics.confusion_matrix(y_test, y_pred))
```
